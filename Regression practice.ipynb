## Split data frame into train and test

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,cross_val_score,RandomizedSearchCV
from sklearn.metrics import mean_absolute_error,confusion_matrix
```

```python
np.random.seed(42)
X=df.drop("MedHouseVal",axis=1)
y=df["MedHouseVal"]

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_test),len(X_train)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
reg=RandomForestRegressor() ## RandomForestRegressor using default parameters value.
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
np.random.seed(42)    
rf_grid={ "n_estimators": np.arange(10,100,10),
          "max_depth": [None,3,5,10],
          "max_features":[0.5,1,"sqrt","auto"],
          "max_samples":[0.8],
          "min_samples_split":np.arange(2,10,2),
          "min_samples_leaf":np.arange(1,20,2)  
} 

rf_model=RandomizedSearchCV(RandomForestRegressor(n_estimators=40,n_jobs=-1),
                            param_distributions=rf_grid,
                            verbose=True,cv=5,n_iter=10)

rf_model.fit(X_train,y_train)
```

```python
reg_new=RandomForestRegressor(max_features='sqrt', max_samples=0.8,
                      min_samples_leaf=7, min_samples_split=8, n_estimators=60,
                      n_jobs=-1)
reg_new.fit(X_train,y_train)
reg_new.score(X_test,y_test)
```

```python
y_pred=reg_new.predict(X_test)
y_pred
```

```python
### Compare prediction value and truth Value

compare={ "Prediction":y_pred,
          "Truth":y_test
    
}

comp=pd.DataFrame(compare)
comp
```
